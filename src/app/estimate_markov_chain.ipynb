```python
PUNCTUATION = ['.', ':', ',', ';', '.', '!', '?', ')', '-', "'", '"', '’']
SENTENCE_END = ['.', '?', '!']
ACCEPTABLE = re.compile(r'[A-Za-z0-9 _\.,!"\'/$\-:\?;\(\))]+')
```

```python
bigram = {}
def bigram_builder(tokens, bigram=bigram):
    for bgr in zip(*[tokens[i:] for i in range(2)]):
        key = bgr[0]
        value = bgr[1]
        if key in bigram:
            bigram[key].append(value)
        else:
            bigram[key] = [value]
```

```python
trigram = {}
def trigram_builder(tokens, trigram, bigram):
    for bgr in zip(*[tokens[i:] for i in range(3)]):
        key = (bgr[0], bgr[1])
        value = bgr[2]
        if not key in trigram:
            # begin with bigram
            trigram[key] = []
            #bigram[bgr[1]]
        # add value twice so that it gets higher weight in randomization
        trigram[key].extend(1*[value])
```

```python
from random import choice
def generate_random_word(begin, trigram):
    return(choice(trigram[begin]))
def punctuation_join(tokens):
    output = ''
    new_sentence = True
    for token in tokens:
        if token == "smiley":
            token = ":-)"
        if new_sentence:
            token = token.title()
            new_sentence = False
        if token in PUNCTUATION:
            if token in SENTENCE_END:
                new_sentence = True
            output += token
        else:
            output += ' '+token
    return output[1:]
```

```python
def generate_random_title(trigram):
    output = []
    words = ('begin1', 'begin2')
    while not words[1] == "end":
        new_word = generate_random_word(words, trigram)
        output.append(new_word)
        words = (words[1], new_word)
    return punctuation_join((output[:-1]))
```

```python
def serialize(trigram):
    ''' json does not allow tuples as keys
     serialize as {"first": {
                         "second1": ["third1", "third2"],
                         "second2": ["third3", "third4"]}}
    '''
    output = {}
    for word1, word2 in trigram.keys():
        if word1 in output:
            output[word1][word2] = trigram[(word1, word2)]
        else:
            output[word1] = {word2: trigram[(word1, word2)]}
    return output
```

```python
import json
serialized = serialize(trigram)
with open('../assets/data/trigrams.json', 'w', encoding='utf8') as f:
    f.write(json.dumps(serialized, ensure_ascii=False))
```
